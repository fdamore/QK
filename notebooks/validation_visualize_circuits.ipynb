```python
from matplotlib import scale
from numpy import size
from pqk.Circuits import Circuits
from qiskit.circuit.library import ZZFeatureMap

x_encode = Circuits.x_encoded(3, full_ent=False)
display(x_encode.draw('mpl', scale=1, fold=-1))

xyx_encode = Circuits.xyz_encoded(3)
display(xyx_encode.draw('mpl', scale=1, fold=-1))

zfm = Circuits.zzfeaturemap(3) #ZZFeatureMap(feature_dimension=3, parameter_prefix='phi')
display(zfm.decompose().draw('mpl', scale=1, fold=-1))

iqp = Circuits.IQP_HuangE2(3)
display(iqp.decompose().draw('mpl', scale=1, fold=-1))

trotter = Circuits.Trotter_HuangE3(3, T=1)
display(trotter.decompose().draw('mpl', scale=1, fold=-1))    
```

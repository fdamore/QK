Data in env.sel3.csv is not scaled
Data in env.sel3.scaled.csv are scaled using standard scaling

In this notebook, data scaled using MinMax ([0 pi] or [0,2pi]) scaled is produces and stored in env.sel3.minmax.csv

```python
import numpy as np
import pandas as pd
import csv
from sklearn.preprocessing import MinMaxScaler, StandardScaler

#load not scaled data
#load dataset with panda
#data are scaled outside the notebook
env = pd.read_csv('env.sel3.csv')  
Y = env[['occupancy']]
X = env[['illuminance', 'blinds','lamps','rh', 'co2', 'temp']]

#features = MinMaxScaler(feature_range=(0, np.pi)).fit_transform(X) #produce pi_minmax
#features = MinMaxScaler(feature_range=(0, 1)).fit_transform(X) #produce 2pi_minmax
features = StandardScaler().fit_transform(X) #standard scaler
env_scaled = pd.DataFrame(features, columns=X.columns)
```

```python
#append the label to the scaled dataset
env_scaled.insert(6,Y.columns[0], Y)
env_scaled.head()
```

```python
#salve to csv
env_scaled.to_csv('env.sel3.sk_sc.csv', quoting=csv.QUOTE_STRINGS, index=False)
```
